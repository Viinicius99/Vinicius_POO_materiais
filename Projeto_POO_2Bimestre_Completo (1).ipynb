## PARTE 1 ‚Äì Classe Principal e Objetos

**Tarefas:**
- Definir e criar a classe principal com pelo menos 3 atributos.
- Implementar o m√©todo construtor (`__init__`).
- Criar pelo menos 2 objetos da classe principal.
- Implementar um m√©todo de exibi√ß√£o de dados.

**C√≥digo:**

```python
# üìò Projeto: Requisi√ß√£o de Materiais Escolares - Vinicius
# Turma: 2ATI
# Autor: Vinicius

# ===================== CLASSES ==========================

class RequisicaoVinicius:
    def __init__(self, solicitante, destino, material, quantidade):
        self.__solicitante = solicitante  # Atributo privado
        self.destino = destino
        self.material = material
        self.quantidade = quantidade

    def exibir_dados_Vinicius(self):
        print(f"\nüì¶ Requisi√ß√£o:\n- Solicitante: {self.__solicitante}\n- Destino: {self.destino}\n- Material: {self.material}\n- Quantidade: {self.quantidade}")

    def get_solicitante_Vinicius(self):
        return self.__solicitante

    def set_solicitante_Vinicius(self, novo_nome):
        self.__solicitante = novo_nome

    def verificar_material_Vinicius(self):
        materiais_disponiveis = ['caneta', 'l√°pis', 'papel', 'borracha']
        if self.material.lower() in materiais_disponiveis:
            print("‚úÖ Material dispon√≠vel no estoque.")
        else:
            print("‚ö†Ô∏è Material pode estar indispon√≠vel.")

class RequisicaoUrgenteVinicius(RequisicaoVinicius):
    def __init__(self, solicitante, destino, material, quantidade, prioridade):
        super().__init__(solicitante, destino, material, quantidade)
        self.prioridade = prioridade

    def marcar_como_urgente_Vinicius(self):
        if self.prioridade:
            print("üö® Requisi√ß√£o marcada como URGENTE!")
        else:
            print("‚è≥ Requisi√ß√£o padr√£o.")

    def exibir_dados_Vinicius(self):
        print(f"\nüö® Requisi√ß√£o URGENTE:\n- Solicitante: {self.get_solicitante_Vinicius()}\n- Destino: {self.destino}\n- Material: {self.material}\n- Quantidade: {self.quantidade}\n- Prioridade: {'Alta' if self.prioridade else 'Baixa'}")

# ===================== DADOS E FUN√á√ïES ==========================
usuarios_vinicius = {}  # Armazenamento de usu√°rios
requisoes_vinicius = []  # Lista de requisi√ß√µes

def cadastrar_usuario_Vinicius():
    print("\nüë§ Cadastro de Usu√°rio")
    while True:
        nome = input("Escolha um nome de usu√°rio: ")
        if nome in usuarios_vinicius:
            print("‚ùå Usu√°rio j√° existe. Tente outro.")
        else:
            break

    senha = input("Escolha uma senha: ")
    tipo = ""
    while tipo not in ["operador", "cliente"]:
        tipo = input("Tipo de usu√°rio (operador / cliente): ").lower()
        if tipo not in ["operador", "cliente"]:
            print("‚ö†Ô∏è Tipo inv√°lido.")

    usuarios_vinicius[nome] = {"senha": senha, "tipo": tipo}
    print(f"‚úÖ Usu√°rio '{nome}' cadastrado como '{tipo}'.")

def login_Vinicius():
    print("\nüîê Login")
    usuario = input("Usu√°rio: ")
    senha = input("Senha: ")

    if usuario in usuarios_vinicius and usuarios_vinicius[usuario]["senha"] == senha:
        print(f"‚úÖ Login bem-sucedido! Bem-vindo, {usuario}.")
        return usuarios_vinicius[usuario]["tipo"]
    else:
        print("‚ùå Usu√°rio ou senha incorretos.")
        return None

def adicionar_requisicao_Vinicius():
    print("\n‚ûï Nova Requisi√ß√£o")
    solicitante = input("Nome do solicitante: ")
    destino = input("Destino: ")
    material = input("Material: ")

    while True:
        try:
            quantidade = int(input("Quantidade: "))
            if quantidade <= 0:
                print("‚ö†Ô∏è Quantidade deve ser maior que 0.")
                continue
            break
        except ValueError:
            print("‚ùå Digite um n√∫mero v√°lido.")

    prioridade = input("√â urgente? (s/n): ").lower() == 's'

    if prioridade:
        req = RequisicaoUrgenteVinicius(solicitante, destino, material, quantidade, True)
    else:
        req = RequisicaoVinicius(solicitante, destino, material, quantidade)

    requisoes_vinicius.append(req)
    print("‚úÖ Requisi√ß√£o adicionada com sucesso!")
    req.exibir_dados_Vinicius()

def listar_requisicoes_Vinicius():
    print("\nüìã Lista de Requisi√ß√µes")
    if not requisoes_vinicius:
        print("üì≠ Nenhuma requisi√ß√£o registrada.")
        return

    for i, req in enumerate(requisoes_vinicius):
        print(f"\nüîπ Requisi√ß√£o #{i + 1}")
        req.exibir_dados_Vinicius()

def excluir_requisicao_Vinicius():
    listar_requisicoes_Vinicius()
    try:
        index = int(input("N√∫mero da requisi√ß√£o a excluir: ")) - 1
        if 0 <= index < len(requisoes_vinicius):
            del requisoes_vinicius[index]
            print("üóëÔ∏è Requisi√ß√£o exclu√≠da.")
        else:
            print("‚ùå √çndice inv√°lido.")
    except ValueError:
        print("‚ùå Entrada inv√°lida.")

def editar_requisicao_Vinicius():
    listar_requisicoes_Vinicius()
    try:
        index = int(input("N√∫mero da requisi√ß√£o a editar: ")) - 1
        if 0 <= index < len(requisoes_vinicius):
            nova_quantidade = int(input("Nova quantidade: "))
            requisoes_vinicius[index].quantidade = nova_quantidade
            print("‚úèÔ∏è Quantidade atualizada.")
        else:
            print("‚ùå N√∫mero inv√°lido.")
    except ValueError:
        print("‚ùå Entrada inv√°lida.")

def salvar_requisicoes_Vinicius():
    print(f"üíæ Simulando salvamento... Total de requisi√ß√µes: {len(requisoes_vinicius)}")

# ===================== MENU PRINCIPAL ==========================

def menu_principal_Vinicius():
    while True:
        print("\n==== MENU INICIAL ====")
        print("1 - Cadastrar novo usu√°rio")
        print("2 - Fazer login")
        print("0 - Sair")

        escolha = input("Escolha uma op√ß√£o: ")

        if escolha == "1":
            cadastrar_usuario_Vinicius()
        elif escolha == "2":
            tipo_usuario = login_Vinicius()
            if tipo_usuario:
                menu_usuario_Vinicius(tipo_usuario)
        elif escolha == "0":
            print("üëã Encerrando o sistema.")
            break
        else:
            print("‚ùå Op√ß√£o inv√°lida.")

def menu_usuario_Vinicius(tipo_usuario):
    while True:
        print("\n===== MENU DO USU√ÅRIO =====")
        print("1 - Adicionar Requisi√ß√£o")
        print("2 - Editar Requisi√ß√£o")
        print("3 - Excluir Requisi√ß√£o")
        print("4 - Listar Requisi√ß√µes")
        print("5 - Salvar")
        print("0 - Logout")

        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            if tipo_usuario == "operador":
                adicionar_requisicao_Vinicius()
            else:
                print("üîí Apenas operadores podem adicionar.")
        elif opcao == "2":
            if tipo_usuario == "operador":
                editar_requisicao_Vinicius()
            else:
                print("üîí Apenas operadores podem editar.")
        elif opcao == "3":
            if tipo_usuario == "operador":
                excluir_requisicao_Vinicius()
            else:
                print("üîí Apenas operadores podem excluir.")
        elif opcao == "4":
            listar_requisicoes_Vinicius()
        elif opcao == "5":
            if tipo_usuario == "operador":
                salvar_requisicoes_Vinicius()
            else:
                print("üîí Apenas operadores podem salvar.")
        elif opcao == "0":
            print("üîô Saindo do menu do usu√°rio.")
            break
        else:
            print("‚ùå Op√ß√£o inv√°lida.")

# ===================== EXECU√á√ÉO ==========================

if __name__ == "__main__":
    menu_principal_Vinicius()
```

```python
# Desenvolva aqui o c√≥digo da Parte 5

!pip install mysql-connector-python


import mysql.connector
import getpass

# Conex√£o com o banco de dados
def conectar():
    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="",  # normalmente em branco no XAMPP
        database="escola_vinicius"
    )

# Classe para requisi√ß√µes
class RequisicaoVinicius:
    def __init__(self, solicitante, destino, material, quantidade, prioridade, usuario_id):
        self.solicitante = solicitante
        self.destino = destino
        self.material = material
        self.quantidade = quantidade
        self.prioridade = prioridade
        self.usuario_id = usuario_id

    def salvar_no_banco(self):
        conexao = conectar()
        cursor = conexao.cursor()
        cursor.execute("""
            INSERT INTO requisicoes_vinicius (solicitante, destino, material, quantidade, prioridade, usuario_id)
            VALUES (%s, %s, %s, %s, %s, %s)
        """, (self.solicitante, self.destino, self.material, self.quantidade, self.prioridade, self.usuario_id))
        conexao.commit()
        conexao.close()
        print("‚úÖ Requisi√ß√£o salva com sucesso!")

# Cadastro de usu√°rio
def cadastrar_usuario():
    nome = input("Digite o nome do novo usu√°rio: ")
    senha = getpass.getpass("Digite a senha: ")
    tipo = input("Tipo de usu√°rio (operador/cliente): ").lower()

    conexao = conectar()
    cursor = conexao.cursor()
    try:
        cursor.execute("""
            INSERT INTO usuarios_vinicius (nome, senha, tipo)
            VALUES (%s, %s, %s)
        """, (nome, senha, tipo))
        conexao.commit()
        print("‚úÖ Usu√°rio cadastrado com sucesso!")
    except mysql.connector.errors.IntegrityError:
        print("‚ùå Nome de usu√°rio j√° existe.")
    conexao.close()

# Login
def login():
    nome = input("Nome de usu√°rio: ")
    senha = getpass.getpass("Senha: ")

    conexao = conectar()
    cursor = conexao.cursor(dictionary=True)
    cursor.execute("""
        SELECT * FROM usuarios_vinicius WHERE nome = %s AND senha = %s
    """, (nome, senha))
    usuario = cursor.fetchone()
    conexao.close()

    if usuario:
        print(f"\nüîì Login bem-sucedido como {usuario['tipo']}!\n")
        return usuario
    else:
        print("‚ùå Usu√°rio ou senha inv√°lidos.")
        return None

# Cadastrar nova requisi√ß√£o
def nova_requisicao(usuario):
    solicitante = input("Solicitante: ")
    destino = input("Destino: ")
    material = input("Material: ")
    quantidade = int(input("Quantidade: "))
    prioridade = input("√â urgente? (s/n): ").lower() == 's'

    req = RequisicaoVinicius(solicitante, destino, material, quantidade, prioridade, usuario['id'])
    req.salvar_no_banco()

# Listar requisi√ß√µes
def listar_requisicoes():
    conexao = conectar()
    cursor = conexao.cursor()
    cursor.execute("SELECT * FROM requisicoes_vinicius")
    registros = cursor.fetchall()
    conexao.close()

    if registros:
        print("\nüìÑ Requisi√ß√µes cadastradas:")
        for r in registros:
            print(f"ID: {r[0]} | Solicitante: {r[1]} | Material: {r[3]} | Quantidade: {r[4]} | Urgente: {'Sim' if r[5] else 'N√£o'}")
    else:
        print("üì≠ Nenhuma requisi√ß√£o encontrada.")

# Excluir requisi√ß√£o
def excluir_requisicao():
    listar_requisicoes()
    id_req = int(input("Digite o ID da requisi√ß√£o para excluir: "))
    conexao = conectar()
    cursor = conexao.cursor()
    cursor.execute("DELETE FROM requisicoes_vinicius WHERE id = %s", (id_req,))
    conexao.commit()
    conexao.close()
    print("üóëÔ∏è Requisi√ß√£o exclu√≠da com sucesso.")

# Menu principal
def menu():
    while True:
        print("\nüìò MENU PRINCIPAL")
        print("1. Cadastrar usu√°rio")
        print("2. Login")
        print("0. Sair")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == '1':
            cadastrar_usuario()
        elif opcao == '2':
            usuario = login()
            if usuario:
                menu_usuario(usuario)
        elif opcao == '0':
            print("üëã Encerrando o sistema.")
            break
        else:
            print("‚ùå Op√ß√£o inv√°lida.")

# Menu ap√≥s login
def menu_usuario(usuario):
    while True:
        print("\nüîß MENU DO USU√ÅRIO")
        print("1. Nova requisi√ß√£o")
        print("2. Listar requisi√ß√µes")
        if usuario['tipo'] == 'operador':
            print("3. Excluir requisi√ß√£o")
        print("0. Logout")
        escolha = input("Escolha uma op√ß√£o: ")

        if escolha == '1':
            nova_requisicao(usuario)
        elif escolha == '2':
            listar_requisicoes()
        elif escolha == '3' and usuario['tipo'] == 'operador':
            excluir_requisicao()
        elif escolha == '0':
            print("üîí Logout realizado.")
            break
        else:
            print("‚ùå Op√ß√£o inv√°lida.")

# Rodar o sistema
if __name__ == "__main__":
    menu()
```
